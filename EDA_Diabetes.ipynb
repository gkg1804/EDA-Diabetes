```python
# importing Important Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# importing the data
df = pd.read_csv('/content/diabetes.csv')
df.head()
```

```python
# Getting the shape of the data
print('Shape : ', df.shape)
```

```python
# getting an idea about the duplicate rows
df.duplicated().sum()
```

```python
# np.int64(0) means there are no duplicates in the dataset
```

```python
# getting an idea about the data types
df.info()
```

```python
# All data types of columns looks correct
```

```python
# Getting an idea about the null values
df.isnull().sum().any()
```

```python
# np.False_ means there are no NULL values in the data
```

```python
# Count plot between Pregnancies and Outcome
plt.figure(figsize = (7, 5))
sns.countplot(x = 'Pregnancies', hue = 'Outcome', data = df)
plt.show()
```

```python
'''
Conclusions ->
1. Women, having maximum 6 pregnancies, have lower chances of being diabetic
2. Women, having 1 or 2 pregnancies, have even lower chances of being diabetic
3. Women, having more than 6 pregnancis, have higher chances of being diabetic
4. Women, having more than 13 pregnancies, have almost 100% chances of being diabetic
'''
```

```python
# Count plot between Glucose and Outcome
plt.figure(figsize = (15, 5))
sns.relplot(x = 'Glucose', y = 'BloodPressure', hue = 'Outcome', data = df)
plt.show()

```

```python
'''
Conclusions ->
1. Women, having Glucose less than 125, have lower chances of being diabetic
2. Women, having Glucose more than 125, have higher chances of being diabetic
3. Being diabetic is not dependent on BloodPressure
'''
```

```python
# Count plot between Insulin and Outcome
plt.figure(figsize = (7, 5))
sns.relplot(x = 'Glucose', y = 'Insulin', hue = 'Outcome', data = df)
plt.show()
```

```python
'''
Conclusions ->
1. Women, having Glucose less than 125 and Insulin less than 250, have lower chances of being diabetic
2. Women, having Glucose greater than 125 and Insulin greater than 250, have higher chances of being diabetic
'''
```

```python
# Count plot between BMI and Outcome
plt.figure(figsize = (7, 5))
sns.countplot(x = 'Outcome', hue = 'SkinThickness', data = df)
plt.show()
```

```python
'''
Conclusions ->
1. Women, having thinner skin, have higer chances of being diabetic
2. Women, having thicker skin, have lower chances of being diabetic
'''
```

```python
# Count plot between DiabetesPedigreeFunction and Outcome
plt.figure(figsize = (20, 5))
sns.relplot(x = 'DiabetesPedigreeFunction', y = 'BMI', hue = 'Outcome', data = df)
plt.show()
```

```python
'''
Conclusions ->
1. Women, having DiabetesPedigreeFunction greater than 0.6, have higher chances of being diabetic
'''
```

```python
# Count plot between Age and Outcome
plt.figure(figsize = (15, 7))
sns.countplot(x = 'Age', hue = 'Outcome', data = df)
plt.show()
```

```python
'''
Conclusions ->
1. Women, having age less than 30, have lower chances of being diabetic
2. Women, having age greater than 30, have higher chances of being diabetic
'''
```

```python
# Correlation Analysis
cm = df.corr(numeric_only = True)
plt.figure(figsize = (20, 5))
sns.heatmap(cm, annot = True)
plt.show()
```

```python
'''
Conclusions ->
1. Age and Pregnancies have moderate positive relationship.
2. Glucose and Outcome have moderate positive relationship.
3. Insulin and SkinThickness have moderate positive relationship.
4. BMI and SkinThickness have moderate positive relationship.
'''
```

```python
# Distribution plot
plt.figure(figsize = (12, 8))

plt.subplot(221)
sns.distplot(df['Pregnancies'])

plt.subplot(222)
sns.distplot(df['Glucose'])

plt.subplot(223)
sns.distplot(df['BloodPressure'])

plt.subplot(224)
sns.distplot(df['SkinThickness'])

plt.show()
```

```python
# Distribution plot
plt.figure(figsize = (12, 8))

plt.subplot(221)
sns.distplot(df['Insulin'])

plt.subplot(222)
sns.distplot(df['BMI'])

plt.subplot(223)
sns.distplot(df['DiabetesPedigreeFunction'])

plt.subplot(224)
sns.distplot(df['Age'])

plt.show()
```

```python
df.groupby('Outcome')[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']].mean().T.plot(figsize=(15,5))
```

```python
'''
Conclusions ->
1. The mean of Glucose is comparetively higher in diabetic women than in non-diabetic women.
2. The mean of Insulin is comparetively higher in diabetic women than in non-diabetic women.
'''
```
